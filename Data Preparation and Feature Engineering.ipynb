```python
#import numpy and pandas
import numpy as np
import pandas as pd

from pathlib import Path
```

```python
data_folder = Path("Data")
data_file = data_folder/ "nyc_benchmarking_disclosure_2017_consumption_data.csv"
data = pd.read_csv(data_file)
```

```python
data = data.replace({'Not found': np.nan, ' Not found ' : np.nan})
```

```python
data.head()
```

```python
# Iterate through the columns
for col in list(data.columns):
    # Select columns that should be numeric
    if ('ft²' in col or 'kBtu' in col or 'Metric Tons CO2e' in col or 'kWh' in 
        col or 'therms' in col or 'gal' in col or 'Score' in col or 'kW' in col):
        # Convert the data type to float
        data[col] = data[col].astype(float)
```

```python
# Function for providing a summary of missing values in a DataFrame
# Courtsey: https://stackoverflow.com/a/39734251

def find_missing_values(df):
    '''Function accepts a dataframe and returns summary of missing values'''
    #find missing values & corresponding percentage
    miss_value = df.isnull().sum()
    miss_value_perc = df.isnull().sum() / len(df)
    
    #concatenate the two data frames into a single table
    miss_value_table = pd.concat([miss_value, miss_value_perc], axis = 1 )
    
    #rename columns and remove columns having no null values
    miss_value_table = miss_value_table.rename(
        columns = {0: 'Missing Values', 1 : '% Missing values'})
    
    # Use iloc to create a bool df for columns which have missing values which is used to filter data
    miss_value_table = miss_value_table[
        miss_value_table.iloc[:,0]!=0].round(3).sort_values(
        '% Missing values', ascending = False)
    
    #format and print a summary
    op_text = f'''Input data frame contains {len(df.columns)} columns of which
    {len(miss_value_table)} columns are having missing values'''
    
    print(op_text)
    
    #return the summary of missing values
    return miss_value_table
  
```

```python
miss_value_summary = find_missing_values(data)
miss_value_summary
```

```python
# All columns which are having higher than 60% missing values will be eliminated for data analysis
miss_value_summary = miss_value_summary[miss_value_summary.iloc[:,1] < .6]
print(f'Dropping {len(miss_value_summary.index)} columns from original dataset')
data.drop(columns = list(miss_value_summary.index), inplace = True)
```
